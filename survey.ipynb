```python
import pandas as pd
import os
```

```python
labels = ['Q1', 'Q2', 'Q3', 'Q4']
models = ['gan', 'transformer_1', 'transformer_2']
gan_path = 'gan_by_emotion/'
transformer_1_path = 'transformer_1_by_emotion/'
transformer_2_path = 'transformer_2_by_emotion/'
```

```python
gan_list = []
transformer_1_list = []
transformer_2_list = []
```

```python
# get the file names for each emotion
for label in labels:
    for model in models:
        if model == 'gan':
            for file in os.listdir(os.path.join(gan_path, label)):
                gan_list.append(model + '_' + label + '_' + file)
        elif model == 'transformer_1':
            for file in os.listdir(os.path.join(transformer_1_path, label)):
                transformer_1_list.append(model + '_' + label + '_' + file)
        elif model == 'transformer_2':
            for file in os.listdir(os.path.join(transformer_2_path, label)):
                transformer_2_list.append(model + '_' + label + '_' + file)
```

```python
# save the file names to a df
gan_df = pd.DataFrame(gan_list, columns=['file_name'])
transformer_1_df = pd.DataFrame(transformer_1_list, columns=['file_name'])
transformer_2_df = pd.DataFrame(transformer_2_list, columns=['file_name'])
```
