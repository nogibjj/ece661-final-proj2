```python
SEQUENCE_LENGTH = 100
LATENT_DIMENSION = 1000
BATCH_SIZE = 16
EPOCHS = 100
SAMPLE_INTERVAL = 1
```

```python
from l_transformer_utils import *

gan = GAN(rows=SEQUENCE_LENGTH)   
     
gan.train(epochs=EPOCHS, batch_size=BATCH_SIZE, sample_interval=SAMPLE_INTERVAL, emotion='Q1')

# Save the generator and discriminator models
gan.generator.save("generator_model_Q1.h5")
gan.discriminator.save("discriminator_model_Q1.h5")
```

```python
from l_transformer_utils import * 
from tensorflow.keras.models import load_model
from music21 import stream, note, chord, instrument

generator_model = load_model("generator_model_Q1.h5")
    
# Load the processed notes and get the number of unique pitches
notes = get_notes(emotion='Q1')
n_vocab = len(set(notes))

# Generate new music sequence
for i in range(5):
    generated_music = generate_music(generator_model, LATENT_DIMENSION, n_vocab)
    
    # Create a MIDI file from the generated music
    create_midi(generated_music, filename="generated_music_Q1_" + str(i))
```

```python
from music21 import converter, instrument, note, chord, stream
for i in range(5):
    file = converter.parse('/workspaces/Transformer_GANs_music_generation/generated_music_Q1_' + str(i) + '.mid') 
    file.show('midi')
```
